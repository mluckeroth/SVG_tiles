```python
%load_ext autoreload
%autoreload 2
import tile_tools
from IPython.display import SVG
import svgpathtools as spt
```

## Plot Original File

```python
file = "source/Globe.svg"
tile_tools.get_svg_dimensions(file)
```

```python
SVG(file)
```

## Set Crop Area and Visualize

```python
crop_rect = (200, 150, 2100, 1000) #rect is tuple (x, y, width, length)
crop_path = tile_tools.rect2path(crop_rect)
new_attr = tile_tools.path2attr(crop_path)
#rebuild SVG with red box
paths, attributes, svg_attributes = spt.svg2paths2(file)
#this example file is missing the attributes from the original file, needs to be added before adding crop path
for a in attributes:
    a['fill'] = "none"
    a['stroke'] = "#404040"
    a['stroke-width'] = "3.0"
paths.append(crop_path)
attributes.append(new_attr)
spt.wsvg(paths, attributes=attributes, svg_attributes=svg_attributes, filename='globe_crop_rect.svg')

SVG('globe_crop_rect.svg')
```

```python
svg_attributes
```

## Build a new file with the cropped content

```python
cropped_paths, cropped_attributes, new_svg_attributes = tile_tools.crop_svg(file, crop_path)
for a in cropped_attributes:
    a['fill'] = "none"
    a['stroke'] = "#404040"
    a['stroke-width'] = "3.0"

spt.wsvg(cropped_paths, 
         attributes=cropped_attributes, 
         svg_attributes=new_svg_attributes, filename='globe_crop_1of3.svg')
SVG('globe_crop_1of3.svg')
```
